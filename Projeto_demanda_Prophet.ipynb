## Introdução a demanda

<center><img width="80%" src="https://img.freepik.com/vetores-gratis/analistas-de-demanda-apertando-as-maos-nas-telas-de-laptops-e-planejando-a-demanda-futura-planejamento-de-demanda-analise-de-demanda-ilustracao-de-conceito-de-previsao-de-vendas-digital_335657-2098.jpg?w=1380&t=st=1673631255~exp=1673631855~hmac=f7e2054ec0e0fc279111f1eea1dbfc6add38035cd40d55e0d011001e79199d18"></center>

Demanda é a quantidade de algo ou serviço procurado por consumidores a um preço definido. Pode-se interpretar a demanda como a procura, mas não ao consumo, pois em alguns casos o cliente pode mostrar interesse momentâneo mas não consumir/adquirir de imediato. 

Um dos objetivos de analisar a demanda é que ela pode ser usada para definir o preço dos produtos. Em alguns casos, a demanda é sazonal, ou seja, há uma variação da quantidade do serviço/produto em determinada estação, momento econômico, renda populacional, etc.De acordo com a lei da demanda o preço é inversamente proporcional a quantidade de demanda. 

Alguns tipos de demanda são destaques [3]:

* Negativa: quando o produto/serviço passa a não ser mais consumido igual a tempos anteriores. 
Inexistente: quando determinado  produto/serviço é desconhecido. 
* Latente: quando a procura por determinado produto/serviço que ainda não se encontra no mercado. 
* Declinante: quando o produto/serviço sofre uma queda na demanda. 
* Irregular: quando determinado produto/serviço possui a demanda sazonal, isto é, uma alta ou baixa em algum período específico. 
* Plena: uma demanda ideal ou seja quando determinado produto/serviço é vendido como o esperado. 
* Excessiva: quando a procura por determinado produto/serviço é maior que o esperado. 

Uma previsão de demanda é a possibilidade de prever/estimar a demanda de consumidores para um produto procurando a disposição de um cliente de comprar o produto por um determinado preço. Para realizar a previsão é necessário que haja uma aquisição dos dados das vendas desse produto/serviço e a partir de modelos estatísticos realizar uma previsão.

## 3. Limpeza dos dados

O primeiro passo na limpeza dos dados ser retirar as virgulas presentes nos preços do vinhos.

### Preparando os dados para o prophet

Antes de usar os dados no Prophet é preciso prepará-los, pois é necessário que os dados estejam em um formato específico. 


* Datas em uma coluna chamada `ds`
* Vendas em um coluna chamda `y`

```python
# importando o Prophet
from prophet import Prophet
import logging
logging.getLogger().setLevel(logging.ERROR)
```

```python
# preparando os dados
df_ts = df_clean.groupby('Date', as_index = False)['sales'].sum()
df_ts['Date'] = pd.to_datetime(df_ts['Date'], format = "%Y-%m-%d")
df_ts.index = pd.to_datetime(df_ts['Date'], format = "%Y-%m-%d")
df_ts.drop('Date', axis = 1, inplace = True)

#nomes corretos
df_ts = df_ts.reset_index()
df_ts.columns = ['ds', 'y']
df_ts.head()
```

É pré-requisito para a realização de previsões em séries temporais que ela seja estacionária. 

Uma série temporal estacionária possui **média**, **variância** e **covariância** constante em função do tempo. Visualmente é:

Alguns passos podem ser considerados para avaliar se uma série temporal é estacionária:

* **Observar o comportamento dos gráficos**: é possível verificar visualmente durante a análise exploratória se a série temporal é estacionária. 
* **Testes estatísticos**: testes estatísticos baseados em metodologia ciêntifica que verifica se uma série temporal é estacionária. 
* **Resumo estatísticos dos dados**: as informações estatítiscas dos dados podem mostar um indício se a série temporal é estacionária. 

Neste projeto será usado o teste **Augmented Dickey Fuller (ADF)** para verificar se a série é estacionária. 

O ADF é um dos testes estatísticos mais usados e nele é considerado duas hipóteses:

* A Hipótese Nula ($H_0$) do teste é que a série temporal **não é estacionária**. 
* A Hipótese Alternativa ($H_1$) **rejeita a hipótese nula**.


Para que a ($H_0$) seja rejeitado é levado em conta o *p-value*. É considerado um valor de significância para o *p-value* que pode ser de 0.01 (1%), 0.05 (5%) ou 0.1 (10%). E quanto menor for *p-value* mais evidências para rejeitar a hipótese nula ($H_0$), ou seja, que a série seja estacionária. 


**Aplicando o teste ADF**:

```python
# importar o teste ADF
from statsmodels.tsa.stattools import adfuller

#extrair apenas os valores
X = df_ts.y

#aplicar ADF e imprimir o resultado
result = adfuller(X)
print("Dicke-Fuller Aumentado")
print("Teste Estatístico : {:.4f}".format(result[0]))
print('Valor-p: {:.4f}'.format(result[1]))
print("Valores Críticos:")
for key, value in result[4].items():
  print("\t{}: {:.4f}".format(key, value))
```

O *p-value* apresentou um valor de `0,1533` muito maior que os valores de significância. Isto pode ser uma evidência de não rejetiar a hipótese nula, ou seja, desta série não ser estacionária. Pode-se até rejeitar a hipótese se fosse adotado um valor de confiança de 85%.


### Série estacionária

Para transformar uma série em estacionária é necessário remover a tendência a sanzonalidade dos dados. Uma maneira de realizar isso é aplicação de *log* e assim reduzindo a magnitudo dos valores da série. E então subtrai-se a média móvel de um período em relação ao *log* da série temporal. 

```python
df_ts.index = pd.to_datetime(df_ts['ds'], format = "%Y-%m-%d")
df_ts.drop('ds', axis = 1, inplace = True)
```

```python
# checagem do plot original
ma = df_ts.rolling(24).mean()

fig,ax = plt.subplots(figsize = (8,4))
df_ts.plot(ax=ax, legend=False)
ma.plot(ax=ax, color='r', legend=False)
sns.despine()
plt.tight_layout();
```

```python
# log
df_log = np.log(df_ts)
ma_log = df_log.rolling(24).mean()

fig,ax = plt.subplots(figsize = (8,4))
df_log.plot(ax=ax, legend=False)
ma_log.plot(ax=ax, color='r', legend=False)
sns.despine()
plt.tight_layout();
```

```python
# subtraindo a media do log nos dados
df_sub = (df_log - ma_log).dropna()
ma_sub = df_sub.rolling(24).mean()
std_sub = df_sub.rolling(24).std()

fig,ax = plt.subplots(figsize = (8,4))
df_sub.plot(ax=ax, legend=False)
ma_sub.plot(ax=ax, color='r', legend=False)
std_sub.plot(ax=ax, color='g', legend=False)
sns.despine()
plt.tight_layout();
```

```python
# repetindo ADF

#extrair apenas os valores
X_sub = df_sub.y

#aplicar ADF e imprimir o resultado
result = adfuller(X_sub)
print("Dicke-Fuller Aumentado")
print("Teste Estatístico : {:.4f}".format(result[0]))
print('Valor-p: {:.10f}'.format(result[1]))
print("Valores Críticos:")
for key, value in result[4].items():
  print("\t{}: {:.4f}".format(key, value))
```

Aqui é apresentado que o *p-value* é `0.0000010231`. Um valor muito abaixo do valor adotado como siginifância. Então pode-se considerar que a série está estacionária.

## Aplicação do Prophet

```python
# retomando o formato para o prophet
df_log.reset_index(inplace=True)
```

Antes de realizar a previsão do prophet é necessário realizar 3 etapas.
1. Definir o tempo de predição. Em que neste caso será o período de 30 dias.
2. Instaciar o modelo. 
3. Treinar o modelo. 

```python
# definindo o tamnho de previsão
prediction_size = 30
train_df = df_log[:-prediction_size]
```

```python
# instanciando o modelo
model = Prophet()
```

```python
# treinando o modelo
model.fit(train_df)
```

```python
# fazendo previsões
future = model.make_future_dataframe(periods=prediction_size)

forecast = model.predict(future)

forecast.head()
```

```python
# plot previsões
model.plot(forecast).savefig('st_forecast.png')
```

O gráfico apresenta as previsões realializadas. Os pontos pretos correspondem aos valores reais dos dados de treino e a linha azul escura os valores previstos. 

```python
# componentes
model.plot_components(forecast).savefig('st_components.png')
```

Aqui apresenta a tendência das vendas anualmente, semanalmente e mensalmente.  

É necéssario verficar agora as métricas dos resultados. Assim vemos o quanto distante está as previsões dos valores reaias. 

Para isso é visto o **erro médio absoluto** (MAE — do inglês Mean Absoluto Error), que mede a média da diferença entre o valor real com o predito. E, o **erro percentual absoluto médio** (MAPE — do inglês Mean Absolute Percentual Error) é uma métrica que mostra a porcentagem de erro em relação aos valores reais.

```python
# reutilizando a função
def make_comparison_dataframe(historical, forecast):
  return forecast.set_index('ds')[['yhat', 'yhat_lower',
                                'yhat_upper']].join(historical.set_index('ds'))
cmp_df = make_comparison_dataframe(df_log, forecast)

# calculando MAPE e MAE
def calculate_forecast_errors(df_ts, prediction_size):

  df = df_ts.copy()
  df['e'] = df['y'] - df['yhat']
  df['p'] = 100 * df['e'] / df['y']
  predicted_part = df[-prediction_size:]

  error_mean = lambda error_name: np.mean(np.abs(predicted_part[error_name]))

  return{'MAPE': error_mean('p'), 'MAE': error_mean('e')}

# imprimindo o MAPE e MAE
for err_name, err_value in calculate_forecast_errors(cmp_df, prediction_size).items():
  print(err_name, err_value)
```

Após verifica-se que os valores de erros estão baixos, é bom observar o resultado completo em gráfico. 

```python
# checando os resultados
fig, ax = plt.subplots(figsize=(12,6))

plt.plot(cmp_df['yhat'])
plt.plot(cmp_df['yhat_lower'])
plt.plot(cmp_df['yhat_upper'])
plt.plot(cmp_df['y'])

plt.xlabel('Tempo')
plt.ylabel('Média de vendas semanais')
ax.plot(cmp_df['yhat'], label='yhat')
ax.plot(cmp_df['yhat_lower'], label='yhat_lower')
ax.plot(cmp_df['yhat_upper'], label='yhat_upper')
ax.plot(cmp_df['y'], label='y')
ax.legend()
plt.show()
```
